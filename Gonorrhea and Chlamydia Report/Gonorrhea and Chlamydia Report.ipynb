```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gonorrhea_df = pd.read_csv('../Datasets/Public Health Gonorrhea Cases 2000-2014.csv', index_col=0)
chlamydia_df = pd.read_csv('../Datasets/Public Health Chlamydia Cases 2000-2014.csv', index_col=0)
```

```python
gonorrhea_df.isna().sum()
```

```python
gonorrhea_df.dropna(inplace=True)
```

```python
gonorrhea_df.sample(5)
```

```python
gonorrhea_df.describe()
```

```python
import matplotlib.pyplot as plt

# Scatter plot for Male 15-44
plt.figure(figsize=(8, 6))
plt.scatter(gonorrhea_df['Year'], gonorrhea_df['Male 15-44'], marker='o')
plt.title('Male 15-44 vs. Year')
plt.xlabel('Year')
plt.ylabel('Male 15-44')
plt.show()

# Scatter plot for Incidence Rate Lower CI
plt.figure(figsize=(8, 6))
plt.scatter(gonorrhea_df['Year'], gonorrhea_df['Incidence Rate Lower CI'], marker='o')
plt.title('Incidence Rate Lower CI vs. Year')
plt.xlabel('Year')
plt.ylabel('Incidence Rate Lower CI')
plt.show()

# Scatter plot for Incidence Rate Upper CI
plt.figure(figsize=(8, 6))
plt.scatter(gonorrhea_df['Year'], gonorrhea_df['Incidence Rate Upper CI'], marker='o')
plt.title('Incidence Rate Upper CI vs. Year')
plt.xlabel('Year')
plt.ylabel('Incidence Rate Upper CI')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the sum of values across the years
sum_values = gonorrhea_df.groupby('Year').sum()

# Create a combined chart with a bar chart and a line chart
plt.figure(figsize=(10, 6))

# Bar chart for the sum of values
plt.bar(sum_values.index, sum_values['Male 15-44'], label='Male 15-44')
plt.bar(sum_values.index, sum_values['Incidence Rate Lower CI'], label='Incidence Rate Lower CI', alpha=0.7)
plt.bar(sum_values.index, sum_values['Incidence Rate Upper CI'], label='Incidence Rate Upper CI', alpha=0.7)

# Line chart to connect the sums of values
plt.plot(sum_values.index, sum_values['Male 15-44'], color='blue', marker='o', linestyle='-', linewidth=2, markersize=7)
plt.plot(sum_values.index, sum_values['Incidence Rate Lower CI'], color='green', marker='o', linestyle='-', linewidth=2, markersize=7)
plt.plot(sum_values.index, sum_values['Incidence Rate Upper CI'], color='red', marker='o', linestyle='-', linewidth=2, markersize=7)

# Customize the chart
plt.title('Sum of Values Across Years (Combined Chart)')
plt.xlabel('Year')
plt.ylabel('Sum')
plt.legend()
plt.grid(True)

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a list of unique years
unique_years = gonorrhea_df['Year'].unique()

# Create box plots for each column across the years
for column in ['Male 15-44', 'Incidence Rate Lower CI', 'Incidence Rate Upper CI']:
    data_to_plot = [gonorrhea_df[gonorrhea_df['Year'] == year][column] for year in unique_years]

    plt.figure(figsize=(10, 6))
    plt.boxplot(data_to_plot, labels=unique_years)
    plt.title(f'Box Plot for {column} Across Years')
    plt.xlabel('Year')
    plt.ylabel('Values')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```
